# Censo de Población y Vivienda (CPV) 2010


## Limpieza de datos


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iter_31_cpv2010.csv')
df
```

```python
df = df[(df['mun'] != 0) & (df['loc'] != 0)]
df
```

```python
df.info()
```

```python
municipios = df[['entidad', 'mun', 'nom_mun', 'pobtot', 'tvivparhab', 'vivtot', 'vph_snbien']]
municipios
```

```python
municipios.info()
```

## Conversión de datos


```python
municipios = municipios.replace("*", 0)
municipios
```

```python
# Convertir columnas a numéricas
for col in ["pobtot", "tvivparhab", "vivtot", "vph_snbien"]:
    municipios[col] = pd.to_numeric(municipios[col], errors="coerce")
```

```python
# Agrupar a nivel municipio
municipios = municipios.groupby(["mun", "nom_mun"]).agg({
    "pobtot": "sum",
    "tvivparhab": "sum",
    "vivtot": "sum",
    "vph_snbien": "sum"
}).reset_index()
```

```python
municipios["personas_por_vivienda"] = municipios["pobtot"] / municipios["tvivparhab"]
municipios["viviendas_por_persona"] = municipios["vivtot"] / municipios["pobtot"]
municipios["porc_sin_bienes"] = municipios["vph_snbien"] / municipios["vivtot"] * 100
municipios['saturacion_viv'] = municipios["pobtot"] / municipios["vivtot"]
municipios
```

## Índice de saturación de vivienda

En este análisis se busca entender por qué algunos municipios de Yucatán
presentan alta o baja saturación de viviendas, es decir, la cantidad promedio de
personas que viven en cada casa, identificando patrones de hacinamiento en los
diferentes municipios


### 10 municipios con mayor saturación


```python
max_saturacion = municipios.nlargest(10, 'saturacion_viv')
max_saturacion[['nom_mun', 'saturacion_viv']]
```

### 10 municipios con menor saturación


```python
min_saturacion = municipios.nsmallest(10, 'saturacion_viv')
min_saturacion[['nom_mun', 'saturacion_viv']]
```

### Gráficas


```python
# Concatenar Top 5 y Bottom 5 en un solo DataFrame
top5 = municipios.nlargest(5, 'saturacion_viv')
bottom5 = municipios.nsmallest(5, 'saturacion_viv')
comb = pd.concat([top5, bottom5])

# Gráfica combinada
plt.figure(figsize=(10,6))
colors = ['tomato']*5 + ['skyblue']*5  # Rojo para top, azul para bottom

plt.bar(comb['nom_mun'], comb['saturacion_viv'], color=colors)
plt.title("  Tops municipios según índice de saturación de vivienda")
plt.ylabel("Personas por vivienda")
plt.xticks(rotation=45)

# Etiquetas encima de las barras
for i, v in enumerate(comb['saturacion_viv']):
    plt.text(i, v + 0.05, f"{v:.2f}", ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Boxplot de todos los municipios
plt.figure(figsize=(8,5))
plt.boxplot(municipios['saturacion_viv'], vert=False, patch_artist=True,
            boxprops=dict(facecolor="lightblue", color="navy"),
            medianprops=dict(color="red", linewidth=2),
            whiskerprops=dict(color="black"),
            capprops=dict(color="black"),
            flierprops=dict(marker='o', color='orange', alpha=0.7))

plt.title("Índice de saturación de vivienda en municipios de Yucatán")
plt.xlabel("Personas por vivienda")
plt.show()
```

### Hipótesis


En los municipios con alta saturación de vivienda, como Tahdziú, Chikindzonot,
Uayma o Chemax, viven más de 4 personas por casa en promedio. Esto puede pasar
porque en estas zonas rurales las familias suelen ser grandes y muchas veces
varias generaciones comparten la misma vivienda. La falta de recursos para
construir nuevas casas y la costumbre de vivir en familia extensa hacen que haya
hacinamiento, lo que puede afectar la calidad de vida dentro de los hogares.

En cambio, en municipios con baja saturación, como Telchac Puerto, Yobaín,
Dzemul o Progreso, hay 2 personas o menos por casa. Esto puede deberse a que son
lugares turísticos o costeros donde hay muchas casas de descanso o inversión que
casi siempre están vacías. También influye que mucha gente migra a Mérida u
otras ciudades, dejando más casas libres de las que realmente se usan.
